```python
dummies = pd.get_dummies(df['improve_yourself_how'], dtype=float)
```

```python
df_new= pd.concat([df, dummies], axis=1)
```

```python

cols= list(dict(df_new['bodyweight'].value_counts()).keys())
vals= []
for i in range(len(cols)):
    vals.append(i)
val_dict = dict((key,val) for key,val in zip(cols,vals))
df_new['bodyweight'] = df_new['bodyweight'].map(val_dict)
```

```python
def preprocess(col):
    cols= list(dict(df_new[col].value_counts()).keys())
    vals= []
    for i in range(len(cols)):
        vals.append(i)
    val_dict = dict((key,val) for key,val in zip(cols,vals))
    df_new[col] = df_new[col].map(val_dict)
```

```python
df_new
```

```python
cols = ['gender','sexuallity','virgin','prostitution_legal','pay_for_sex','race']
for col in cols:
    preprocess(col)
```

```python
import lightgbm as lgb
X = df_new.drop(['attempt_suicide'],1)
y = df_new['attempt_suicide']
```

```python
X=  X.drop(['job_title'],1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
import xgboost as xgb
```

```python
xgb_params = {
    'object': 'binary:hinge:',
    'eval_metric ': 'auc',
    'max_depth': 20,
    'max_bin': 200,
    'num_iterations': 1000,
    'learning_rate': 0.01
}

dtrain = xgb.DMatrix(X_train,y_train)
dtest = xgb.DMatrix(X_test,y_test)

gbd2 = xgb.train(xgb_params,dtrain,evals=[(dtrain,'train'),(dtest,'valid')])
```

```python
gbd2.predict(xgb.DMatrix(X_test))
```

```python
lgb_params = {
    'object': 'binary',
    'metric': 'auc',
    'max_depth': 20,
    'max_bin': 200,
    'num_iterations': 1000,
    'learning_rate': 0.01
}

dtrain = lgb.Dataset(X_train,y_train)
dtest = lgb.Dataset(X_test,y_test)

gbd = lgb.train(lgb_params,dtrain,valid_sets=dtest)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
param = {
    'bagging_freq': 5,
    'bagging_fraction': 0.4,
    'boost_from_average':'false',
    'boost': 'gbdt',
    'feature_fraction': 0.05,
    'learning_rate': 0.01,
    'max_depth': -1,  
    'metric':'auc',
    'min_data_in_leaf': 80,
    'min_sum_hessian_in_leaf': 10.0,
    'num_leaves': 13,
    'num_threads': 8,
    'tree_learner': 'serial',
    'objective': 'binary', 
    'verbosity': 1
}
folds = StratifiedKFold(n_splits=10, shuffle=False, random_state=44000)
oof = np.zeros(len(X))
feature_importance_df = pd.DataFrame()

for fold_, (trn_idx, val_idx) in enumerate(folds.split(X.values, y.values)):
    print("Fold {}".format(fold_))
    trn_data = lgb.Dataset(X.iloc[trn_idx], label=y.iloc[trn_idx])
    val_data = lgb.Dataset(X.iloc[val_idx], label=y.iloc[val_idx])

    num_round = 1000000
    clf = lgb.train(param, trn_data, num_round, valid_sets = [trn_data, val_data], verbose_eval=1000, early_stopping_rounds = 3000)
    oof[val_idx] = clf.predict(X.iloc[val_idx], num_iteration=clf.best_iteration)
    
    fold_importance_df = pd.DataFrame()
    
    fold_importance_df["importance"] = clf.feature_importance()
    fold_importance_df["fold"] = fold_ + 1
    feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
    
    


print("CV score: {:<8.5f}".format(np.mean(y== oof)))
```
